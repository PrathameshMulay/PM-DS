# Remove outlier

```python
#Standardisation on continuous data  is performed in the form of feature scaling    (x-mu/s)
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X1=pd.DataFrame(ss.fit_transform(A[con]),columns=con)
outlier=X1[(X1.SalePrice>3)|(X1.SalePrice<-3)]          #we need to remove outlier from training data for better prediction
                                                          #so we take house price which are too great than mean or too less after standardization
```

```python
outlier
```

```python
ol=[]         #create an empty list
for i in con:
    ol.extend(outlier.index)  #join list of indexes of outliers....there are same values for multiple cols so we need to use unique
from numpy import unique
ol=unique(ol)
```

```python
X=X.drop(index=ol,axis=0)       
Y=Y.drop(index=ol,axis=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X1=pd.DataFrame(ss.fit_transform(X[con]),columns=con)
X2=pd.get_dummies(X[cat])
Xnew=X1.join(X2)
Xnew.shape
```

```python
for i in Xnew.columns:
    if Xnew[i].dtypes=="object":
        x=Xnew[i].mode()[0]
        Xnew[i]=Xnew[i].fillna(x)
    else:
        x=Xnew[i].mean()
        Xnew[i]=Xnew[i].fillna(x)
```

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

```python
#OLS method
from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ol = OLS(ytrain,xconst)
model = ol.fit()
rsq = model.rsquared_adj
col_to_drop = model.pvalues.sort_values().index[-1]
print("Dropped: column:",col_to_drop,"\tRsquared:",round(rsq,4))
Xnew = Xnew.drop(labels=col_to_drop,axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
xconst = add_constant(xtrain)
ol = OLS(ytrain,xconst)
model = ol.fit()
rsq = model.rsquared_adj
```

```python
adj_R2=[]
Q=range(0,99,1)
```

```python
count=0
for i in Q:
    cols_to_drop=model.pvalues.sort_values().index[-2]          #index[-1] is constant we added above which cant be removed as it is B0 in eqn
    Xnew=Xnew.drop(labels=cols_to_drop,axis=1)
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
    from statsmodels.api import OLS,add_constant
    xconst=add_constant(xtrain)
    ols=OLS(ytrain,xconst)
    model=ols.fit()
    adj_R2.append(model.rsquared_adj)
    count+=1
    print(count)
```

# LINEAR MODEL

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(xtrain,ytrain)
pred_tr=model.predict(xtrain)
pred_ts=model.predict(xtest)
from sklearn.metrics import mean_absolute_error
tr_err=mean_absolute_error(ytrain,pred_tr)
ts_err=mean_absolute_error(ytest,pred_ts)
```

```python
tr_err,ts_err                       ##test_error>>train_error    thus we need to do regularization (ridge or lasso)
```

```python
lambdas = []
q = 8
for i in range(0,4000,1):
    q = q + 0.001
    q = round(q,4)
    lambdas.append(q)
```

```python
from sklearn.linear_model import Ridge
tr = []
ts = []
for i in lambdas:
    rr = Ridge(alpha=i)
    model = rr.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err = mean_absolute_error(ytrain,tr_pred)
    ts_err = mean_absolute_error(ytest,ts_pred)
    tr.append(tr_err)
    ts.append(ts_err)
```
